Task 1 - subtask 2

```python
#Task 1 - 2
size = 100000

# Creating array
exp_data = npy.random.exponential(scale=1.0, size=(size, 1))
uniform_data = npy.random.uniform(low=0, high=5, size=(size, 1))
normal_data = npy.random.normal(loc=2.5, scale=1.0, size=(size, 1))

# Setting up the plot
fig, ax = plt.subplots(figsize=(12, 8))

# Plot histograms
bins = 100  # Increased number of bins for better resolution
ax.hist(exp_data, bins=bins, alpha=0.5, color = 'red' , label='Exponential', density=True)
ax.hist(uniform_data, bins=bins, alpha=0.5, color = 'blue', label='Uniform', density=True)
ax.hist(normal_data, bins=bins, alpha=0.5, color = 'green', label='Normal', density=True)

# labeling
ax.set_xlabel('Value')
ax.set_ylabel('Density')
ax.set_title('Distribution Comparison: Exponential, Uniform, and Normal')
ax.legend(loc = "upper left")
ax.axis([-1, 6, 0, 1])

# Show the plot
plt.tight_layout()
plt.show()
```

Task 1 - subtask 3

```python

from mpl_toolkits.mplot3d import Axes3D


# Creating a grid of x and y values
x = npy.linspace(-5, 5, 100)
y = npy.linspace(-5, 5, 100)
X, Y = npy.meshgrid(x, y)

# Calculating Z values
Z = X**2 + Y**2

# Creating the 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plotting the surface
surface = ax.plot_surface(X, Y, Z, cmap='viridis')

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Z = X^2 + Y^2')

# Adding a color bar
fig.colorbar(surface, ax=ax, shrink=0.5, aspect=5)

# Show the plot
plt.show()
```

```python
import seaborn as sns
import pandas as pd

df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')
features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']

df.head()

# Pearson correlation
pearson_corr = df[features].corr(method='pearson')
plt.figure(figsize=(10, 8))
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Pearson Correlation Coefficient')
plt.show()
```

```python

# Spearman correlation
spearman_corr = df[features].corr(method='spearman')
plt.figure(figsize=(10, 8))
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Spearman Rank Correlation')
plt.show()
```
